```python
# Linear Regression between Height and weight

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score

y = df['Weight']
x = df['Height']

plt.plot(x,y,'ro')
b1,b0 = np.polyfit(x,y,deg=1)
print('y hat = ',b0,'+',b1,'x')

plt.plot(x,y,'go')
b1,b0 = np.polyfit(x,y,deg=1)
yp = b0 + b1*x
plt.plot(x,yp)
plt.show()
# Coefficient of determination
print(r2_score(y,yp))
plt.scatter(y,x,color = 'goldenrod')
b1,b0 = np.polyfit(y,x,1)
y_p = b0 + b1*np.array(y)
plt.plot(y, y_p)
plt.show()
residuals = np.array(df.Height - y_p)
plt.scatter(y_p, residuals , color = 'red')
plt.plot([0,200],[0,0])
plt.show()
```

```python
y_2008_bro = df[df['Year']==2008][df[df['Year']==2008]['Medal']=='Bronze']
y_2008_bro = pd.concat([y_2008_bro,pd.get_dummies(y_2008_bro.Medal)],axis = 1).groupby(['Team'],as_index=False).sum()
plt.barh(y_2008_bro.Team,y_2008_bro.Bronze)
plt.title("No of Medals won by each country in 2008")
plt.xlabel('Bronze MEDALS in 2008')
plt.ylabel('Countries')
plt.show()
```

```python
groupedGender = pd.concat([df,pd.get_dummies(df.Sex)],axis=1).groupby(['Team','ID'],as_index = False).sum()
groupedGender
```

```python
groupedGender = groupedGender.groupby('Team',as_index=False).sum()
groupedGender
```

```python
plt.bar(groupedGender.Team,groupedGender.M)
plt.bar(groupedGender.Team,groupedGender.F)
plt.legend(['Male','Female'])
plt.xlabel("Team")
plt.ylabel("PARTICIPANTS")
plt.xticks(rotation =-90)
plt.title("Male and  Female Participants over the years")
```

```python
Russia = df[df['Team']=='Russia']
Russia = pd.concat([Russia,pd.get_dummies(Russia.Medal)],axis = 1).groupby(['Year'],as_index = False).sum()
Russia
width = 0.80
plt.bar(Russia.Year - 0.8 ,Russia.Bronze,width,color = 'blue')
plt.bar(Russia.Year + 0 , Russia.Silver,width,color = 'Red')
plt.bar(Russia.Year +0.8, Russia.Gold,width,color = 'orange')
plt.legend(['Bronze','Silver','Gold'])
plt.title('Medals Won by Russia')
plt.xlabel('Year')
plt.ylabel(' Number of Medals')
plt.plot()
```
