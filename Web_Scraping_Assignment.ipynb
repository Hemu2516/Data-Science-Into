```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import requests as r

url = 'https://www.amazon.in/Samsung-Galaxy-Prime-Light-128GB/dp/B0BD3V985M/ref=sr_1_1_sspa?crid=3DOMW883V78J7&keywords=samsung%2Bm32&qid=1668443249&qu=eyJxc2MiOiIzLjcxIiwicXNhIjoiMy4zNiIsInFzcCI6IjIuOTcifQ%3D%3D&sprefix=sam%2Caps%2C260&sr=8-1-spons&sp_csd=d2lkZ2V0TmFtZT1zcF9hdGY&th=1'
```

```python
res = r.get(url)
data = bs(res.content,'html.parser')
res
```

```python
data
```

```python
N1= data.find_all(class_ = 'a-profile-name')
Names=[]
for  i in N1:
    Names.append(i.get_text())
```

```python
Names
```

```python
R1 = data.find_all(class_ = 'a-icon-alt')
Ratings = []
for i in range(1,len(R1)):
    Ratings.append(R1[i].get_text())
```

```python
Ratings
```

```python
D0 = data.find_all(class_='a-size-base review-title a-text-bold')
Desc=[]
for i in D0:
    Desc.append(i.string)
```

```python
D1 = data.find_all(class_='a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold')

for i in D1:
    Desc.append(i.span.string)
```

```python
Desc
```

```python
len(Names),len(Ratings),len(Desc)
```

```python
import pandas as pd
```

```python
Dataset = pd.DataFrame()
Dataset['Reviewer'] = Names[:5]
Dataset['Ratings']  = Ratings[:5]
Dataset['Description'] = Desc[:5]
```
