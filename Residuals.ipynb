```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math 
import statistics
```

```python
df = pd.read_csv('Residuals.csv')
df
```

```python
#x =wind speed
#y =Current

plt.scatter(df['Wind Speed'], df.Current,color = 'goldenrod')
b1,b0 = np.polyfit(df['Wind Speed'],df.Current,1)

y_p = b0 + b1*np.array(df['Wind Speed'])
plt.plot(df['Wind Speed'], y_p)
plt.show()
```

```python
residuals = np.array(df.Current - y_p)
plt.scatter(y_p, residuals , color = 'red')
plt.plot([1,4],[0,0])
plt.show()
```

```python
# ln(x) =wind speed
#y =Current

plt.scatter(np.log(df['Wind Speed']), df.Current,color = 'goldenrod')
b1,b0 = np.polyfit(np.log(df['Wind Speed']),df.Current,1)

y_p = b0 + b1*np.array(np.log(df['Wind Speed']))
plt.plot(np.log(df['Wind Speed']), y_p)
plt.show()
residuals = np.array(df.Current - y_p)
plt.scatter(y_p, residuals , color = 'red')
plt.plot([0,4],[0,0])
plt.show()
```
