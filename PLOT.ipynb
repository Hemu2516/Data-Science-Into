```python
# ScatterPlot
x = [1 ,2, 3 ,4]
y = [1 ,4 ,9 ,16]
x = [i for i in range(21)]
y =[i*i for i in x]
plt.scatter(x,y,color = 'b' ,marker ='s')
plt.ylabel('dependent variable')
plt.xlabel('independent variable')
plt.show()
```

```python
x = [i for i in range(21)]
y =[i*i for i in x]
b = [i*3 for i in x]

plt.scatter(x,y ,marker ='s',alpha = 0.5)
plt.scatter(x,b,x,color = 'r',)
plt.ylabel('dependent variable')
plt.xlabel('independent variable')
plt.show()
```

```python
#Vertical Bar Graph

x = [1,2,3,4]
height = [2,4,5,3]
labels = ['cars','bikes', 'cycles', 'pedestrians']
y = np.arange(0.2,100)
plt.bar(x,height ,align ='center')
plt.xticks(x,labels)
plt.yticks(x) #Optional to set the values of y
plt.show()
```

```python
# Horizontal 
x = [1,2,3,4]
height = [2,4,5,3]
labels = ['cars','bikes', 'cycles', 'pedestrians']
y = np.arange(0.2,100)
plt.barh(x,height ,align ='center')
plt.xticks(x)
plt.yticks(x,labels) #Optional to set the values of y
plt.show()
```

```python
#edge aligned bar graphs
plt.bar(x,height ,align ='edge',color ='goldenrod')
plt.xticks(x,labels)
#plt.yticks(x,labels) #Optional to set the values of y
plt.show()
```

```python
# Stacked Bar Chart 
x = [1 ,2 ,3 ,4]
men= [4 ,3,8,10]
women = [2,5 ,9, 10]
labels = ['athletics', 'aquatics','hockey','tennis']
p1 = plt.bar(x,men)
p2 = plt.bar(x,women,bottom=men)
plt.xticks(x,labels)
plt.legend((p1[0],p2[0]),('Men','Women'))
plt.show()
```

```python
# Grouped Bar Chart
x = np.arange(5)
y1 = [34 ,56,12,89,67]
y2 = [12,56,78,45,90]
width = 0.30

# plot data in grouped manner of bar type
plt.bar(x-0.15 , y1,width)
plt.bar(x+0.15 , y2,width)
plt.legend((p1[0],p2[0]),('Y1','Y2'))
plt.show()
```

```python
# Histogram 
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.xlim(40, 160)
plt.ylim(0, 0.03)
plt.grid(True)
plt.show()
```

```python
import matplotlib.mlab as mlab
#example data
mu=100
sigma=15
x=mu+sigma*np.random.randn(10000)
num_bins=20
#histogram
n,bins,patches=plt.hist(x,num_bins,color='orange',edgecolor='black')
```

```python
y=mlab.normpdf(bins,mu,sigma)
plt.plot(bins,y,'r--')
plt.xlabel('Smarts')
plt.ylabel('Probs')
plt.title('Histogram')
plt.sublplots_adjust()
plt.show()
```
